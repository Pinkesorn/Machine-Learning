# Perceptron Scikit-learn [AND]

```python
import numpy as np
```

```python
X = [[0,0],
     [0,1],
     [1,0],
     [1,1]]
```

```python
type(X)
```

```python
X_train = np.array(X)
X_train
```

```python
y = [0,0,0,1] # target กรณี AND Gate
```

```python
y_train = np.array(y)
y_train
```

```python
from sklearn.linear_model import Perceptron

model = Perceptron(tol=1e-3,random_state=0,verbose= True) # create Neuron
model.fit(X_train, y_train) # สั้ง train
```

```python
print('score:',model.score(X_train,y_train)) # ตรวจสอบ score
print('predict:',model.predict(X_train))
print('target:',np.array(y))
```

```python
print('weight. [w1 w2] = ',model.coef_[0])
print('w0 (bias) = ',model.intercept_[0])
```

# XOR Gate ด้วย MLP

```python
import numpy as np #MLP
X = [[0,0],
     [0,1],
     [1,0],
     [1,1]]

X_train = np.array(X)
print(X_train)
print()
y = [0,1,1,0] # target กรณี XOR Gate
y_train = np.array(y)
print(y_train)
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(4),random_state=1,verbose=True) # สร้าง NN
model.fit(X_train,y_train)
```

```python
print('score:',model.score(X_train,y_train)) # ตรวจสอบ score
print('predict:',model.predict(X_train))
print('target:',np.array(y))
```

```python
model.max_iter
```

```python
model = MLPClassifier(hidden_layer_sizes=(4),random_state=1,verbose=True,max_iter=850) # สร้าง NN
model.fit(X_train,y_train)
```

```python
print('score:',model.score(X_train,y_train)) # ตรวจสอบ score
print('predict:',model.predict(X_train))
print('target:',np.array(y))
```

```python
[coef.shape for coef in model.coefs_] # ดูโครงสร้าง (shape)
```

# MLP จำเเนกผลไม้ 2 คลาส

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.target.unique()
```

```python
sns.scatterplot(x='m',y='d',hue='target',data=df,style='target',s=140)
plt.show()
```

```python
X = df.drop('target',axis=1)
y = df.target
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X)
y_train = y
```

```python
df.sample(6,random_state=1)
```

```python
df_sc = pd.DataFrame(X_train,columns=['d','m'])
df_sc['target'] = y_train
df_sc.sample(6,random_state=1)
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(random_state=1,verbose=True) # randomstate กำหนดให้สุ่มได้ค่าเหมือนเดิมตลอด
model.fit(X_train,y_train)
```

```python
print('score:', round(model.score(X_train, y_train), 4)) # ตรวจสอบ score
print('actual y:', np.array(y))
print('predicted:', model.predict(X_train))
```

```python
from sklearn.metrics import confusion_matrix,classification_report

y_predict = model.predict(X_train)
print('Score -> {:.4f}'.format(model.score(X_train, y_train)))
print(confusion_matrix(y_train, y_predict))
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(4.5, 3.5))
plot_decision_regions(X_train,np.array(y_train), clf=model, legend=2)
# Get the number of hidden layers from the model
hidden_layer_sizes = model.hidden_layer_sizes
# If hidden_layer_sizes is an integer (one hidden layer), use it directly
# Otherwise, assume it's a tuple and use the first element (size of the first hidden layer)
hidden = hidden_layer_sizes if isinstance(hidden_layer_sizes, int) else hidden_layer_sizes[0]
plt.title('MLP: 2 Class (hidden layer: {})'.format(hidden))
plt.xticks([])
plt.yticks([])
plt.xlabel('m(gram)')
plt.ylabel('d (diameter)')
plt.show()
```

```python
[coef.shape for coef in model.coefs_] # ดูโครงสร้าง (shape)
```

## เตรียมข้อมูล

```python
X = df.drop('species',axis=1)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_sc = sc.fit_transform(X)
X_sc[:5]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_sc,y,test_size=0.25,random_state=1)
len(X_train),len(X_test)
```

## Train and Test

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(4),max_iter= 2000,random_state=1,verbose= True) # randomstate กำหนดให้สุ่มได้ค่าเหมือนเดิมตลอด
model.fit(X_train,y_train)
```

```python
print('score:', round(model.score(X_train, y_train), 4)) # ตรวจสอบ score
print('actual y:', np.array(y))
print('predicted:', model.predict(X_train))
print()
print([coef.shape for coef in model.coefs_])
```
